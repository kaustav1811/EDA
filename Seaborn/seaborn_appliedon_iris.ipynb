```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
import pandas as pd

numeric_df = df.select_dtypes(include = ['number'])

correlation_matrix = numeric_df.corr()

print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,annot = True,cmap = 'coolwarm')
```

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = df, kind = "hex")
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = df, kind = "hex")
```

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = df, kind = "reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = "species") #this will show graphs related to species
```

```python
sns.distplot(df['sepal_width'])
```

```python
sns.distplot(df['sepal_width'], kde = False, bins = 10)
```

```python
sns.countplot(data = df, x = "species")
```

```python
df['species'].value_counts()
```
